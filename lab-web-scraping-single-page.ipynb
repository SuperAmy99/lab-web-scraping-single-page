```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# find url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# download html with a get request
response = requests.get(url)
response.status_code
```

```python
# parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# check that the html code looks like it should
# soup
```

```python
#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p
#chart-position-2 > div.chart-content.col-xs-12.col-sm-8 > p
```

```python
# find the title selector
soup.select('div.chart-content.col-xs-12.col-sm-8 > p > cite')
```

```python
# find the artist selector
# soup.select('div.chart-content.col-xs-12.col-sm-8 > p > em')
```

```python
title = []
for t in soup.select('div.chart-content.col-xs-12.col-sm-8 > p > cite'):
    title.append(t.get_text())
print(title)
len(title)
```

```python
artist = []
for a in soup.select('div.chart-content.col-xs-12.col-sm-8 > p > em'):
    artist.append(a.get_text())
print(artist)
len(artist)
```

```python
songlist = pd.DataFrame({"title":title,
                       "artist":artist,
                      })
songlist
```

```python
url_bill = 'https://www.billboard.com/charts/hot-100/'
```

```python
response = requests.get(url_bill)
response.status_code
```

```python
soup_bill = BeautifulSoup(response.content, "html.parser")
```

```python
bill_list = []

for e in soup_bill.find_all(attrs={'class':'o-chart-results-list-row-container'}):
    bill_list.append({
        'title':e.h3.get_text(strip=True),
        'artist':e.h3.find_next('span').get_text(strip=True)
    })

bill_list
```

```python
billboard_df = pd.DataFrame(bill_list)
billboard_df
```

```python
hot_list = pd.concat([songlist,billboard_df], axis = 0)
hot_list
```
